```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("gopay_me.jpg")
cv2.imshow("gambar", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img) #karena BGR, sehingga berubah
```

```python
img_rgb = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB) #convert ke RGB
plt.imshow(img_rgb)
```

```python
len(img_rgb)
```

```python
img_gray = cv2.cvtColor(img.copy(), cv2.COLOR_BGR2GRAY) #convert ke GRAY
plt.imshow(img_gray, cmap="gray")
```

```python
img_gray
```

```python
ret, thresh = cv2.threshold(img_gray, 185, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresh, cmap='gray')
```
