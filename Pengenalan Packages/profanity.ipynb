```python
import time, random

def start():
    print("Let's get started!")
    kesempatan = int(input("You're about to solve this in: "))
    to_range = int(input("Guessing 1 to "))
    print(f"I had selected a number between 1 and {to_range}. Guess that wisely!")
    number = random.randint(1, to_range)
    while True:
        tebakan = int(input())
        if tebakan == number:
            print("Congratulation!")
            print("Selamat tinggal") if input("Apakah anda ingin mengulang game kembali (y/n)? ").lower() != "y" else start()
            break
        else:
            kesempatan -= 1
            if tebakan > number:
                print("Terlalu besar!")
            else:
                print("Terlalu kecil!")
            if kesempatan > 1:
                print(f"Anda masih memiliki {kesempatan} kesempatan!")
            elif kesempatan == 1:
                print(f"Ini kesempatan terakhirmu!")
            else:
                print(f"Anda kalah! Angka rahasiamu adalah {number}")
                print("Selamat tinggal") if input("Apakah anda ingin mengulang game kembali (y/n)? ").lower() != "y" else start()
                break

start()
```
