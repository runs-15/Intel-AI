```python
data_karyawan = pd.read_csv('[Dataset]_Module11_Train_(Employee).csv')
data_karyawan
```

```python
data_karyawan = data_karyawan.dropna()
data_karyawan.info()
```

```python
data_input = data_karyawan[['Age', 'Education_Level', 'Time_of_service', 'Time_since_promotion', 'growth_rate', 'Travel_Rate', 'Post_Level', 'Pay_Scale', 'Work_Life_balance', 'VAR1', 'VAR2', 'VAR3', 'VAR4', 'VAR5', 'VAR6', 'VAR7']]
data_output = data_karyawan['Attrition_rate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
```

```python
model = LinearRegression()
model_lasso = Lasso()
```

```python
model.fit(x_train, y_train)
model_lasso.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred_lasso = model_lasso.predict(x_test)
```

```python
y_pred
```

```python
y_pred_lasso
```

## Evaluasi Model


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print(mean_absolute_error(y_pred, y_test))
print(mean_absolute_error(y_pred_lasso, y_test))
```

```python
print(mean_squared_error(y_pred, y_test))
print(mean_squared_error(y_pred_lasso, y_test))
```
